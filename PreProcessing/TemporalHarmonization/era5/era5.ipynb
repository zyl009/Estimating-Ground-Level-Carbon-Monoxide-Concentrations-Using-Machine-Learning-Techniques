```python
import xarray as xr
import pandas as pd
import numpy as np
```

```python
era5=xr.open_dataset("/path/to/era5_2019_2024.nc")
# Print CRS
print("CRS:", era5.rio.crs)
```

```python
era5
```

```python
era5= era5.rio.write_crs("EPSG:4326", inplace=True) 
```

```python
print(era5.rio.crs)
```

```python
era5_df=era5.to_dataframe()
era5_df.reset_index(drop=False, inplace=True)
era5_df
```

```python
era5_df['valid_time'] = pd.to_datetime(era5_df['valid_time'])  # Ensure datetime format
era5_df.rename(columns={'valid_time':'time'},inplace=True)
era5_df
```

```python
era5_df.drop(columns={'number','spatial_ref'},inplace=True)
```

```python
# Wind Speed Calculation
era5_df['wind_speed'] = np.sqrt(era5_df['u10']**2 + era5_df['v10']**2)

# Wind Direction Calculation
era5_df['wind_dir'] = (270 - np.degrees(np.arctan2(era5_df['v10'], era5_df['u10']))) % 360
era5_df
```

```python
era5_df.drop(columns={'u10','v10'},inplace=True)
```

```python
era5_df
```

```python
lai = pd.read_feather('/path/to/lai.feather')

# View the dataset
lai
```

```python
lai['time'] = pd.to_datetime(lai['time'])  # Ensure datetime format
```

```python
lai.drop(columns={'number','step','surface','valid_time'},inplace=True)
lai
```

```python
ds=lai.merge(era5_df,on=['time','latitude','longitude'],how='left')
ds
```

```python
# convert time zone
ds['time'] = ds['time'].dt.tz_localize('UTC')      
ds['time'] = ds['time'].dt.tz_convert('Europe/Rome')
ds
```

```python
#set index
ds = ds.set_index('time')

# 1. Current Day Value (21:00 previous day to 15:00 current day)
# Create mask for current window
current_mask = (ds.index.hour >= 21) | (ds.index.hour <=15)
current_data = ds[current_mask].copy()

# Adjust grouping dates for current window:
current_data['group_date'] = current_data.index.where(
    current_data.index.hour <= 15,  # Keep times <15:00 as current date
    current_data.index + pd.Timedelta(days=1)  # Shift >=21:00 to next day
).floor('D')

# Group by adjusted date and location
current_grouped = current_data.groupby([
    'group_date',
    'latitude',
    'longitude'
]).mean().reset_index()

# 2. Previous Value (15:00-21:00 of previous day)
# Create mask for previous window
pre_mask = (ds.index.hour > 15) & (ds.index.hour < 21)
pre_data = ds[pre_mask].copy()

# Group by original date and location
pre_grouped = pre_data.groupby([
    pd.Grouper(freq='D'),
    'latitude',
    'longitude'
]).mean().reset_index()

# Shift previous values forward by 1 day to align with current
pre_grouped['group_date'] = pre_grouped['time'] + pd.Timedelta(days=1)
pre_grouped = pre_grouped.drop(columns=['time'])

# 3. Combine and Clean
# Merge current and previous values
combined = pd.merge(
    current_grouped,
    pre_grouped,
    on=['group_date', 'latitude', 'longitude'],
    suffixes=('', '_pre')
)

# Trim edge dates (first and last day may have incomplete data)
combined = combined[
    (combined['group_date'] > combined['group_date'].min()) &
    (combined['group_date'] < combined['group_date'].max())
]

# Final cleanup
combined = combined.rename(columns={'group_date': 'date'})
```

```python
combined['date']=combined['date'].dt.date
combined
```

```python
combined.to_feather('/path/to/era5_21_15.feather')
```
